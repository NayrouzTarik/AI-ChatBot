```python
import PyPDF2
from nltk.tokenize import sent_tokenize

# Function to extract text from a PDF file
def extract_text_from_pdf(pdf_path):
    text = ""
    with open(pdf_path, 'rb') as file:
        reader = PyPDF2.PdfReader(file)
        for page_num in range(len(reader.pages)):
            text += reader.pages[page_num].extract_text()
    return text

# Function to chunk text based on sentences
def chunk_text(text):
    sentences = sent_tokenize(text)
    chunks = []
    current_chunk = ""
    for sentence in sentences:
        if len(current_chunk) + len(sentence) < 500:  # Example chunk size limit
            current_chunk += " " + sentence
        else:
            chunks.append(current_chunk.strip())
            current_chunk = sentence
    if current_chunk:
        chunks.append(current_chunk.strip())
    return chunks

# Path to the PDF file
pdf_path = 'infos/pdf2.pdf'

# Extract text from the PDF
pdf_text = extract_text_from_pdf(pdf_path)

# Chunk the extracted text
text_chunks = chunk_text(pdf_text)

# Save chunks to a file and print each chunk on a new line
with open('chunked.txt', 'w', encoding='utf-8') as f:
    for chunk in text_chunks:
        f.write(chunk + '\n')
        print(chunk)
```
